```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=7)
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
R = LinearRegression(normalize=False)
R.fit(X_train,y_train)

print("Score of the train set",R.score(X_train,y_train))
print("Score of the test set",R.score(X_test,y_test))
```

```python
r_sq = R.score(X_train, y_train)
print('Coefficient of determination (R2):', r_sq)
```

```python
print('intercept:', R.intercept_)
print('slope:', R.coef_)
```

```python
def adj_r2 (X,y,model):
    r_squared = model.score(X,y)
    return(1 - (1-r_squared)*(len(y)-1)/(len(y)-X.shape[1]-1))
```

```python
print("Adj. R2 of the train set",adj_r2(X_train,y_train,R))
print("Adj. R2 of the test set",adj_r2(X_test,y_test,R))
```

```python
#The Features Coeff
importance = R.coef_
for i in range(len(importance)):
    print( df_bstn.columns[i], "Score:", importance[i])
```
